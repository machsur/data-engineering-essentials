## Joining Data in Pandas Dataframes
Here are the steps involved in joining data using Pandas Dataframe APIs.
* First, we need to define index for both the dataframes that are supposed to be joined.
* Use join to join the Dataframes based on index. The default join is left, we need to make sure right join is used based on the requirement.

In our case, we will join `orders` and `customers` using inner join.

```python
import json
```

```python
import pandas as pd
```

```python
def get_column_names(schemas, ds_name, sorting_key='column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    return [col['column_name'] for col in columns]
```

```python
schemas = json.load(open('data/retail_db/schemas.json'))
```

```python
orders_columns = get_column_names(schemas, 'orders')
```

```python
orders = pd.read_csv(
    'data/retail_db/orders/part-00000',
    names=orders_columns
)
```

```python
customers_columns = get_column_names(schemas, 'customers')
```

```python
customers_columns
```

```python
customers = pd.read_csv(
    'data/retail_db/customers/part-00000',
    names=customers_columns
)
```

```python
customers.set_index('customer_id')
```

```python
customers = customers.set_index('customer_id')
```

```python
orders = orders.set_index('order_customer_id')
```

```python
customer_orders = customers. \
    join(orders, how='inner')
```

```python
customer_orders. \
    reset_index(names='customer_id'). \
    groupby('customer_id')['customer_id']. \
    agg(order_count='count'). \
    reset_index(). \
    query('order_count >= 10')
```
